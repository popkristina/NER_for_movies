```python
with open("../../predictions/bert_large_cased_best_matched_format_2_TRAIN.json") as json_file:
    bert_large_2 = json.load(json_file)
```

```python
bert_large = dict()
for key in bert_large_2.keys():
    if key not in bert_large.keys():
        bert_large[key] = dict()
    for pred in bert_large_2[key]:
        tmp = pred.split(" ")
        if tmp[0] == "-mp":
            if "positive_movies" not in bert_large[key].keys():
                bert_large[key]["positive_movies"] = []
            bert_large[key]["positive_movies"].append(tmp[1][1:-1])
        if tmp[0] == "-mn":
            if "negative_movies" not in bert_large[key].keys():
                bert_large[key]["negative_movies"] = []
            bert_large[key]["negative_movies"].append(tmp[1][1:-1])
        if tmp[0] == "-kp":
            if "positive_keywords" not in bert_large[key].keys():
                bert_large[key]["positive_keywords"] = []
            bert_large[key]["positive_keywords"].append(tmp[1][1:-1])
        if tmp[0] == "-kn":
            if "negative_keywords" not in bert_large[key].keys():
                bert_large[key]["negative_keywords"] = []
            bert_large[key]["negative_keywords"].append(tmp[1][1:-1])   
        if tmp[0] == "-gp":
            if "positive_genres" not in bert_large[key].keys():
                bert_large[key]["positive_genres"] = []
            bert_large[key]["positive_genres"].append(tmp[1][1:-1])
        if tmp[0] == "-gn":
            if "negative_genres" not in bert_large[key].keys():
                bert_large[key]["negative_genres"] = []
            bert_large[key]["negative_genres"].append(tmp[1][1:-1])
        if tmp[0] == "-amp":
            if "positive_actors" not in bert_large[key].keys():
                bert_large[key]["positive_actors"] = []
            bert_large[key]["positive_actors"].append(tmp[1][1:-1])
        if tmp[0] == "-amn":
            if "negative_actors" not in bert_large[key].keys():
                bert_large[key]["negative_actors"] = []
            bert_large[key]["negative_actors"].append(tmp[1][1:-1])
```

```python
import nltk
from nltk.tag.stanford import StanfordNERTagger
# Imports
import os
import sys
import json
import pandas as pd
import string
import numpy as np
import re
import string
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer

import spacy
import en_core_web_lg
from nltk import FreqDist

nlp = en_core_web_lg.load()
```

## 1. SpaCy

```python
os.chdir('D:/TU_Graz/Thesis/Datasets/Reddit_preprocessed')
submissions = pd.read_csv("test_submissions_simplified.csv")
submissions_tokenized = pd.read_csv("test_submissions_tokenized.csv")
sentence_nums = [int(sent.split(" ")[1]) for sent in submissions_tokenized["Sentence"]]
submissions_tokenized["Sentence_index"] = pd.Series(sentence_nums)
```

```python
subs = submissions[['id','text']]
for i in range(0, len(subs['text'])):
    tmp = subs.iloc[i]['text']
    subs.iloc[i]['text'] = re.sub(r'[^\w\s]', '', tmp)
    tmp = subs.iloc[i]['text'] 
    subs.iloc[i]['text'] = re.sub('NEW_LINE', '', tmp)
    tmp = subs.iloc[i]['text'] 
    subs.iloc[i]['text'] = re.sub('new_line', '', tmp)
    
all_movies = dict()
```

```python
for index, row in subs.iterrows():
    doc = nlp(row['text'])
    if row['id'] not in all_movies.keys():
        all_movies[row['id']] = dict()
    if 'positive_movies' not in all_movies[row['id']].keys():
        all_movies[row['id']]['positive_movies'] = list()
    for word in doc.ents:
        if word.label_ in ['WORK_OF_ART', 'ORG']:
            all_movies[row['id']]['positive_movies'].append(word.text)
        print(word.text, word.label_)
```

## 1. SpaCy

```python
os.chdir('D:/TU_Graz/Thesis/Datasets/Reddit_preprocessed')
submissions = pd.read_csv("train_submissions_simplified.csv")
submissions_tokenized = pd.read_csv("train_submissions_tokenized.csv")
sentence_nums = [int(sent.split(" ")[1]) for sent in submissions_tokenized["Sentence"]]
submissions_tokenized["Sentence_index"] = pd.Series(sentence_nums)
```

```python
subs = submissions[['id','text']]
for i in range(0, len(subs['text'])):
    tmp = subs.iloc[i]['text']
    subs.iloc[i]['text'] = re.sub(r'[^\w\s]', '', tmp)
    tmp = subs.iloc[i]['text'] 
    subs.iloc[i]['text'] = re.sub('NEW_LINE', '', tmp)
    tmp = subs.iloc[i]['text'] 
    subs.iloc[i]['text'] = re.sub('new_line', '', tmp)
    
all_actors = dict()
```

```python
for index, row in subs.iterrows():
    doc = nlp(row['text'])
    if row['id'] not in all_actors.keys():
        all_actors[row['id']] = dict()
    if 'positive_actors' not in all_actors[row['id']].keys():
        all_actors[row['id']]['positive_actors'] = list()
    for word in doc.ents:
        if word.label_ == 'PERSON':
            all_actors[row['id']]['positive_actors'].append(word.text)
        print(word.text, word.label_)
```

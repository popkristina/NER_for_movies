```python
# Imports

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_data():
    os.chdir('D:/TU_Graz/Thesis/Datasets/Reddit_features')
    train = pd.read_csv("train_final_all.csv")
    test = pd.read_csv("test_final_all.csv")
    data = train.append(test)
    return data
```

```python
#Words tagged as B-org
data.loc[data['BIO'] == 'B-movie-pos', 'Token'].head()
```

```python
#Words distribution across Tags
plt.figure(figsize=(15, 5))
ax = sns.countplot('BIO', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="center")
plt.tight_layout()
plt.show()
```

```python
#Words distribution across Tags without O tag
plt.figure(figsize=(15, 5))
ax = sns.countplot('BIO', data=data.loc[data['BIO'] != 'O'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="center")
plt.tight_layout()
plt.show()
```

```python
#Words distribution across POS
plt.figure(figsize=(15, 5))
ax = sns.countplot('POS_tag', data=data, orient='h')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="center")
plt.tight_layout()
plt.show()
```

```python
# Read the data
print("Loading the data...")
data = read_data()
print("Done.")
```

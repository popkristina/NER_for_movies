```python
textstat.set_lang('en')
```

```python
#Check the readability of the submissions
#More readability scores availble at: https://pypi.org/project/textstat/

def text_stats_and_readability_scores(text):
    syllable_count = textstat.syllable_count(text)
    number_of_words = textstat.lexicon_count(text, removepunct=True)
    number_of_sentences = textstat.sentence_count(text)
    
    flesch_score = textstat.flesch_reading_ease(text)
    flesch_kincaid = textstat.flesch_kincaid_grade(text)
    automated_readability_index = textstat.automated_readability_index(text)
    
    res = [syllable_count, number_of_words, number_of_sentences, flesch_score, flesch_kincaid, automated_readability_index]
    return res
    
        
def avg_text_stats_and_scores(sentences):
    
    n = len(sentences)
    sum_syl = 0
    sum_num_words = 0
    sum_num_sent = 0
    sum_flesch = 0
    sum_flesch_kinc = 0
    sum_automated_rdb = 0
    
    for line in sentences:
        stats = text_stats_and_readability_scores(line)
        sum_syl += stats[0]
        sum_num_words += stats[1]
        sum_num_sent += stats[2]
        sum_flesch += stats[3]
        sum_flesch_kinc += stats[4]
        sum_automated_rdb += stats[5]
        
    avg_syllable_count = sum_syl/n
    avg_num_words = sum_num_words/n
    avg_num_sentences = sum_num_sent/n
    avg_flesch_score = sum_flesch/n
    avg_flesch_kincaid = sum_flesch_kinc/n
    avg_automated_rdb_index = sum_automated_rdb/n
    
    res = [avg_syllable_count, avg_num_words, avg_num_sentences, avg_flesch_score, avg_flesch_kincaid, avg_automated_rdb_index]
    return res


def plot_histograms_for_readability_scores(sentences):
    x_flesch = []
    x_kincaid = []
    x_automated = []
    
    for line in sentences:
        res = text_stats_and_readability_scores(line)
        x_flesch.append(res[3])
        x_kincaid.append(res[4])
        x_automated.append(res[5])
        
    plt.style.use('ggplot')
    plt.title('Flesch Reading Ease Score')
    plt.xlabel('scores')
    plt.ylabel('frequencies')
    plt.hist(x_flesch, bins = 20)
    plt.show()
    
    plt.title('Flesch-Kincaid Grade Level Score')
    plt.xlabel('scores')
    plt.ylabel('frequencies')
    plt.hist(x_kincaid, bins = 20)
    plt.show()
    
    plt.title('Automated Readability Index')
    plt.xlabel('scores')
    plt.ylabel('frequencies')
    plt.hist(x_automated, bins = 20)
    plt.show()
```

```python
#Print the histograms for the readability scores for the Submission Titles
#They generally have high readability
plot_histograms_for_readability_scores(submissions['reddit_submission_title'])
```

```python
#Average stats for the Submission Title

stats_and_redability_scores_title = avg_text_stats_and_scores(submissions['reddit_submission_title'])
print("Average number of syllables: ", stats_and_redability_scores_title[0])
print("Average number of words: ", stats_and_redability_scores_title[1])
print("Average number of sentences: ", stats_and_redability_scores_title[2])
print("Average Flesch Reading Ease score: ", stats_and_redability_scores_title[3])
print("Average Flesch Kincaid score: ", stats_and_redability_scores_title[4])
print("Average Automated readability index: ", stats_and_redability_scores_title[5])
```

```python
#Print the histograms for the readability scores for the Submission Texts
#We can notice that the texts are not as readable as the titles 

submissions['reddit_submission_text'] = submissions['reddit_submission_text'].replace(np.nan, "_")
plot_histograms_for_readability_scores(submissions['reddit_submission_text'])
```

```python
#Average stats for the Submission Text

stats_and_redability_scores_text = avg_text_stats_and_scores(submissions['reddit_submission_text'])
print("Average number of syllables: ", stats_and_redability_scores_text[0])
print("Average number of words: ", stats_and_redability_scores_text[1])
print("Average number of sentences: ", stats_and_redability_scores_text[2])
print("Average Flesch Reading Ease score: ", stats_and_redability_scores_text[3])
print("Average Flesch Kincaid score: ", stats_and_redability_scores_text[4])
print("Average Automated readability index: ", stats_and_redability_scores_text[5])
```

```python
#Plot a histogram of the movie title lengths in words so that we can know in what range we expect our movie names

title_lengths = []
for line in movie_titles['title']:
    splits = line.split(' ')
    splits = splits[:-1]
    title_lengths.append(len(splits))
    
plt.title('Frequency of movie title lengths')
plt.hist(title_lengths)
plt.show()

print("The maximal title length consists of: ", max(title_lengths), " tokens")
#We see that most movies are short-titles, only a minor percentage have titles that vary from 6 to 12 words.
```
